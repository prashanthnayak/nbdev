- Can be done mentally or with mala/finger counting
- Maintains satva, continuous learning connection, and parampara flow

## Main Teaching: What is Jyotish?

**Jyotish = Study of 33 Devas (not just planets/signs)**

### 1. **Ashtavasu (8 Vassus/Lights)**
- **Sun & Moon** - Original light givers (luminaries)
- **5 Tatwa Planets** - Give knowledge through their physical nature:
  - **Mars** (Agni) - Fire, energy, engineering
  - **Mercury** (Prithvi) - Earth, solid state
  - **Jupiter** (Akasha) - Space, magnetism, connection
  - **Venus** (Jala) - Water, liquid state
  - **Saturn** (Vayu) - Air, movement, breath, longevity
- **Nakshatras** - Mental desires and emotional states

### 2. **Dvadasha Adityas (12 Adityas)**
- The 12 months/signs representing **all material resources**
- Sun's movement creates 30-degree portions (rashis)
- Each sign = time period + space
- **Vishnu Aditya** youngest

### 3. **Ekadasha Rudra (11 Rudras)**
- Address **destruction energies** in horoscope
- 10 hold prana (life force) in body
- 11th relates to individual soul (calculated from Atmakaraka)
- Connected to **Rudra Bhavas** (houses that can end life)

### 4. **Indra & Prajapati (2)**
- **Prajapati** = Lagna/1st house (our creation/birth)
- **Indra** = 10th house (throne, control of senses, karma)

## Key Concepts Clarified

**Rashis vs Bhavas:**
- **Rashis** = Material resources, physical quantities ("heap")
- **Bhavas** = Our thoughts/perceptions about life aspects

**Important Axes:**
- **1st-7th**: Birth-Death, Me-Others (horizontal, eye-level)
- **4th-10th**: Midnight-Noon, Lowest-Highest point (vertical)

**Three Material Worlds:**
- **Bhuloka** - Earth (where we are)
- **Bhuvarloka** - Solar system (Sun, planets)
- **Svargaloka** - Sky (nakshatras, constellations)
