## Summary of Lesson 4: Panchanga (Five Limbs of Time)

## Opening Context
- This is part of the **Aramba (preparatory) portion** of PJC Year 1
- Aramba introduces many concepts that students will revisit throughout the year
- Focus is on understanding basic concepts before diving deeper

## What is Panchanga?

**Pancha** = Five, **Anga** = Limbs

The five fundamental divisions of time that form the **foundation of any horoscope**. Panchanga must be examined **before** looking at planetary positions, as it shows the **quality of time** for that birth.

### The Five Limbs:

1. **Vara** (Day) - Agni Tatva
2. **Tithi** (Lunar day) - Jala Tatva  
3. **Karana** (Half-tithi) - Prithvi Tatva
4. **Nakshatra** (Lunar mansion) - Vayu Tatva
5. **Yoga** (Sun-Moon combination) - Akasha Tatva

## Benefits of Learning Panchanga

- **Vara** - Prolongs life
- **Tithi** - Brings prosperity
- **Karana** - Gives success in all tasks
- **Nakshatra** - Removes sin
- **Yoga** - Immunity from disease

## 1. VARA (The Day)

**Connection**: Agni (fire), energy levels, physical health

**Calculation**: Sunrise to next sunrise

**Seven Varas**: Ravivara (Sunday), Somavara (Monday), Mangalavara (Tuesday), Buddhavara (Wednesday), Guruvara (Thursday), Shukravara (Friday), Shanivara (Saturday)

**Analysis**: 
- Check **Varesha** (lord of the day) placement from **Lagna**
- Also check Varesha placement from **Surya**

**Example Given**: Franklin D. Roosevelt
- Born Monday (Varesha = Moon)
- Moon in 10th house conjunct Mars
- Moon has **zero Digbala** (directional strength) in 10th
- Moon-Mars = Gandanta (fire-water conflict)
- Moon in 6th from Sun (enemy position)
- Result: Polio (health issues related to movement on 4th-10th axis)

### Vara Chakra Concepts:
- **6th graha** = enemy (Shatru)
- **8th graha** = eclipse (Grahana) - blocks manifestation

## 2. TITHI (Lunar Day)

**Connection**: Jala Tatva, relationships, emotional content, interactions

**Calculation**: Sun-Moon relationship; 1 tithi = 12 degrees of zodiac

**30 Tithis total**: 15 in Krishna Paksha (waning), 15 in Shukla Paksha (waxing)

**Eight Lords**: Surya, Chandra, Mangala, Budha, Guru, Shukra, Shani, Rahu (Ketu has no lordship)

**Analysis**: Check Tithi lord placement from **4th house**

### Problematic Tithis:

**Rikta Tithis** (4th, 9th, 14th) - Most problematic:
- Too much emotion, self-centered
- **Chaturthi** - Ganesha (deity)
- **Navami** - Durga (deity)
- **Chaturdashi** - Shiva (deity)

**Other Doshas**:
- **Ekadashi** (11th) - Addiction/anger issues (Mars/Agni)  fasting advised
- **Trayodashi** (13th) - Wealth problems
- **Krishna Chaturdashi** & **Shukla Pratipada** - Serious relationship issues (related to Soma's sin with Tara)

**Example Given**: Marilyn Monroe
- Tithi lord = Venus
- Venus in 10th house in Aries (enemy sign)
- Zero Digbala
- Had relationships but they were problematic

## 3. DIG CHAKRA (Directional Strength)

Eight directions where grahas have **full potential**:

- **East** (Indra) - **Surya** - Self-realization
- **Southeast** (Agni) - **Shukra** - Details, art
- **South** (Yama) - **Mangala** - Warrior energy
- **Southwest** (Niriti) - **Rahu** - Chaos
- **West** (Varuna) - **Shani** - Detachment
- **Northwest** (Vayu) - **Chandra** - Movement
- **North** (Kubera) - **Budha** - Wealth, profession
- **Northeast** (Ishana) - **Guru** - Knowledge

Plus: **Upward** (Ketu/Ananta - Moksha), **Downward** (Lagna/Brahma - Rebirth)

## 4. KALA CHAKRA (Time/Blockage Wheel)

Shows which grahas **block** each direction:

- Mangala blocks Shukra (Agni corner)
- Guru blocks Mangala (Yama corner)
- Budha blocks Rahu (Niriti corner)
- Shukra blocks Shani (Varuna corner)
- Chandra blocks Budha (Kubera corner)
- Rahu blocks Guru (Ishana corner)
- Ketu blocks Surya (Indra corner)

**Key Point**: These are **pairs** that create either debate (Sattvic) or war (Tamasic)

- Cannot make final judgments from Panchanga alone
- Must learn grahas, rashis, bhavas before complete analysis
- Return to Aramba concepts throughout the year as understanding deepens

## Summary of Lesson 6: Katapayadi & Bhava Chakras

## 1. KATAPAYADI SYSTEM (Numerical Values from Names)

**Core Principle:** Sanskrit consonants have numerical values (0-9)

**Key Rules:**
- **Vowels = 0** (only counted if name starts with vowel)
- Use **pronunciation**, not spelling
- Drop incomplete consonants (without following vowel)
- **Ra/La rule:** If consonants joined with R or L, keep only initial consonant
- **Reverse the number** after calculation
- **Divide by 12** to get remainder = Rashi

**Example: Vivekananda**
- Consonants: Va-Ve-Ka-Na-Na-Da
- Values: 4-4-1-0-8
- Number: 44108  Reversed: 80144
- After dividing by 12  Remainder = **4** (karka rahsi)

**Usage:**
- **Rashi** (sign) = interaction with people
- **House number** = inner manifestation

---

## 2. ASTRONOMICAL BASICS

**Horizon:** Where earth and sky meet (observer-dependent)

**Key Points:**
- **Udaya Lagna** = Eastern horizon (rising point)
- **Asta Lagna** = Western horizon (setting point) - 180 from Udaya
- **MC (Mid-Heaven/Zenith)** = 90 above observer
- **Ecliptic** = Horizontal plane between Sun & Earth

**Directional Degrees:**
- North = 0/360
- East = 90
- South = 180 
- West = 270

---

## 3. BHAVA CHAKRA (House System)

**Method:** Take Lagna degree  15 degrees = 30 house span

**Example: Swami Vivekananda (26 Dhanu Lagna)**
- 1st House = 11 Dhanu to 11 Makara
- 2nd House = 11 Makara to 11 Kumbha
- And so on...

**Result:** Planets may **shift houses** if they're within 15 of cusp

**In Vivekananda's chart:**
- Mars shifted from 5th  4th house
- Venus shifted from 2nd  1st house
- Jupiter shifted to 10th house

**Significance:** When many planets shift = **major life transformations**

---

## 4. BHAVA CHALIT CHAKRA

**Starting Point:** Uses **Bhava Lagna** (special lagna, moves 2 hours/sign)

**Same process:** Bhava Lagna degree  15

**Key Difference:** 
- Bhava Chakra starts from **Udaya Lagna**
- Bhava Chalit starts from **Bhava Lagna** (calculated separately)

---

## 5. BIOLOGICAL CLOCK CHART (Practical Application)

**Setup:**
1. Draw Bhava Chalit Chakra
2. Place **6 AM** where Sun is located
3. Place **6 PM** in 7th house from Sun
4. Each house = **2-hour period**

**Planet Activities & Best Times:**

| Planet | Activity | Best Time |
|--------|----------|-----------|
| **Surya** | Spiritual practice, energy level | Sun's house + aspects |
| **Chandra** | Eating/food | Moon's house + aspects |
| **Mangala** | Exercise, physical strength | Mars house + aspects (4th, 7th, 8th from itself) |
| **Budha** | Reading, learning, playing | Mercury's house + aspects |

| **Shukra** | Relationships, indulgence | Venus's house + aspects |

| **Rahu** | **NO activity** - meditation time | Rahu's house + aspects |

**Important Notes:**
- Check planet's **aspects** (5th, 7th, 9th, 2nd from itself for Rahu)
- When planets conjunct = divide time between both activities
- Each house = 2 hours, but find at least 15 min for that activity
- Movement is **anti-zodiacal** (Sun moves through houses, not signs)

## 6. CHART TYPES (Three Representations)

1. **Surya Chakra** (East Indian) - for spirituality

## Opening Context

## CORE DEFINITION: What is a Graha?

**Parashara's Definition (Chapter 3, Shloka 1):**
Heavenly bodies classified by:
1. **Visibility** (can we see them?)
2. **Movement** (do they move?)

- **Moving bodies** = **Graha** (from root "grahana" = to seize/eclipse)
- **Fixed bodies** = **Nakshatra** (fixed stars)

**KEY POINT:** Grahas have the inherent nature to **eclipse the mind** - this is their fundamental characteristic.

---

## Why Rahu & Ketu are Grahas (Despite Having No Physical Body)

**Question addressed:** If we only count visible bodies, why include Rahu/Ketu?

**Answer:**
- They are **mathematical points** (nodes of the Moon)
- They create **inverse light/shadow** (like closing eyes after seeing sun = black circle)
- They represent **extremes of the mind**:
  - **Rahu** = Fully material, maximum desire, wants to consume everything
  - **Ketu** = Fully detached, wants nothing, renunciation

**Manifestations:**
- **Rahu:** Foreigners, different cultures, borders, geographical boundaries, fermentation, alcohol, medicine
- **Ketu:** Outcasts, monks/nuns, spiritual people, those who renounce society
- **Example given:** Man who moved to attic for years, refused all contact = clear Ketu influence

---

## PLANETARY MOTION (Critical Concept)

### Three Categories of Motion:

1. **Always Direct (Forward):**
   - **Surya (Sun) & Chandra (Moon)** - the Luminaries
   - Establish dharma, predictable, "one foot after another"

2. **Always Retrograde (Backward):**
   - **Rahu & Ketu** - the Nodes
   - Move opposite to all other planets
   - Cause eclipses when they meet luminaries

3. **Can Be Either (Tara Grahas - 5 Planets):**
   - **Mangala, Budha, Guru, Shukra, Shani**
   - When **direct**  behave like Luminaries (normal, predictable)
   - When **retrograde**  behave like Nodes (unpredictable, unfulfilled desires)

---

## RETROGRADE PLANETS (Vakri)

**Nature:** Planet behaves like Rahu/Ketu

**Characteristics:**
- **Not progressing** - standing still or pacing
- **Unfulfilled desire/karma** with that planet
- **Unpredictable, odd behavior**
- **Cannot be trusted** (unexpected turns)
- **May not give results** (especially in Prashna/horary)

**Strength Implications:**
- Retrograde = **Extremely strong** (like nodes)
- Has **Cheshtabala** (directional strength)
- Can cause **highest grahana** (eclipse power)

**Prashna (Horary) Rules:**
- If significator is vakri  **Won't happen** (or needs 3 attempts)
- **3x multiplier:** 3x more effort, 3 attempts, or 3 blessings/children

---

## EIGHT TYPES OF MOTION

**5 Direct Motions:**
1-5. (Various speeds of forward movement)

**3 Retrograde Motions:**
1. **Vakra** - Standard retrograde
2. **Anu-vakra** - Slowing down in retrograde
3. **Kutila** - **Stationary** (about to change direction)

**IMPORTANT:** Kutila applies to **both** transitions:
- Direct  Retrograde
- Retrograde  Direct

---

## WHEN DO PLANETS GO RETROGRADE?

**Outer Planets (Mars, Jupiter, Saturn):**

**Exercise given:** Draw blank chart, place Jupiter on Lagna

- **120 from Sun** (5th or 9th house) = **Kutila (Stationary)**
  - 5th house = Just entering retrograde
  - 9th house = Just finishing retrograde
- **180 from Sun** (7th house) = **Highest retrograde motion**
- **6th house** = Vakra (Mars nature)
- **8th house** = Anu-vakra (Rahu nature)
- **4th/10th houses** = Direct motion range

**Inner Planets (Mercury, Venus):**
- Requires deep astronomical study (Surya Siddhanta)
- Not explained in basic course

## DO WE MOVE RETROGRADE PLANETS IN THE CHART?

**NO!** Planet stays where it is.

**Example:** Retrograde boyfriend still behaves oddly, but he's still your boyfriend in the same place.

**What changes:**

- Can cancel exaltation/debilitation (taught in next lesson)
- Creates yogas from its actual position

## NAKSHATRA vs RASHI (Critical Distinction)

**Question:** Why do grahas show different colors in nakshatra vs rashi?

**Answer - The Three Lokas:**

1. **Bhuloka (Earth)** - Where we are
   - **Rashis** = All material resources, physical interactions
   
2. **Bhuvarloka (Solar System)** - Above us
   - **Grahas** = Moving planets, casting shadows
   
3. **Svargaloka (Sky)** - Above that
   - **Nakshatras** = Fixed stars, mental/emotional plane, devatas

**How Grahas Work:**
- Grahas move through nakshatras (truth)
- Cast shadow on both nakshatras AND rashis
- **Bind** nakshatra significations with rashi significations

**Nakshatras** = Your **thoughts, emotions, ideas** (lunar, night concept)
- Where graha is = your specific lens of thinking/feeling

**Rashis** = **Material resources, physical things** (everything you can see/touch)
- Where graha is = where you experience grahana materially

**Example: Two planets in same rashi but different nakshatras:**
- Working on same project (rashi)
- But have different ideas/emotions about it (nakshatra)

**Direction of influence works both ways:**
1. Mental  Material: Thoughts manifest as actions
2. Material  Mental: Events cause us to think certain thoughts

## SAM-YOGA (Combinations)

**Sam-yoga** = ALL ways planets can combine (with planets, nakshatras, rashis)

**Sam-bandha** = **Important** combinations that have great impact

- 4 specific types (one being conjunction/yuti)

**Sam-yoga Sphuta** = Mathematical meeting point of two planets
- Calculated even if not in same rashi
- Shows where yoga will manifest

**Later:** Yoga Sphuta can combine ANY two points (e.g., Lagna lord + 7th lord)

## BHAVISHYA vs ATITA YOGA

**Bhavishya Yoga** = Yoga that **will happen in future**
- Planets moving toward conjunction

**Atita/Veeta Yoga** = Yoga that **happened in past**
- Planets moving apart after conjunction

**Why calculate Atita Yoga Sphuta?**
- See where past yoga had impact
- Useful for research/deep study of charts
- Example: Two people with Chandra-Ketu in 11th - did it happen already or will it happen?

**If yoga rashi not visible from yoga:**
- Don't use seasons (mentioned but corrected)
