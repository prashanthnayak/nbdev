## Key Practices Introduced

- Prayer/worship of the Guru (13 stanzas)
- Done first thing after morning bath
- Can be recited in one sitting as daily Guru prayer
- Traditionally sung at initiations in the parampara

These represent the five tattvas and are worshipped in order:

1. **Ganesha** (Prithvi/Earth) - Must be worshipped first; sits in Muladhara chakra

4. **Devi** (Jala/Water) - Brings sustenance and empathy

## Main Teaching: What is Jyotish?

**Jyotish = Study of 33 Devas (not just planets/signs)**

### 1. **Ashtavasu (8 Vassus/Lights)**
- **Sun & Moon** - Original light givers (luminaries)
- **5 Tatwa Planets** - Give knowledge through their physical nature:
  - **Mars** (Agni) - Fire, energy, engineering
  - **Mercury** (Prithvi) - Earth, solid state
  - **Jupiter** (Akasha) - Space, magnetism, connection
  - **Venus** (Jala) - Water, liquid state
  - **Saturn** (Vayu) - Air, movement, breath, longevity
- **Nakshatras** - Mental desires and emotional states

### 2. **Dvadasha Adityas (12 Adityas)**
- The 12 months/signs representing **all material resources**
- Sun's movement creates 30-degree portions (rashis)
- Each sign = time period + space
- **Vishnu Aditya** youngest

### 3. **Ekadasha Rudra (11 Rudras)**
- Address **destruction energies** in horoscope
- 10 hold prana (life force) in body
- 11th relates to individual soul (calculated from Atmakaraka)
- Connected to **Rudra Bhavas** (houses that can end life)

### 4. **Indra & Prajapati (2)**
- **Prajapati** = Lagna/1st house (our creation/birth)
- **Indra** = 10th house (throne, control of senses, karma)

## Key Concepts Clarified

**Rashis vs Bhavas:**
- **Rashis** = Material resources, physical quantities ("heap")
- **Bhavas** = Our thoughts/perceptions about life aspects

**Important Axes:**
- **1st-7th**: Birth-Death, Me-Others (horizontal, eye-level)
- **4th-10th**: Midnight-Noon, Lowest-Highest point (vertical)

**Three Material Worlds:**
- **Bhuloka** - Earth (where we are)
- **Bhuvarloka** - Solar system (Sun, planets)
- **Svargaloka** - Sky (nakshatras, constellations)

## Summary of Lesson 4: Panchanga (Five Limbs of Time)

## Opening Context
- This is part of the **Aramba (preparatory) portion** of PJC Year 1
- Aramba introduces many concepts that students will revisit throughout the year
- Focus is on understanding basic concepts before diving deeper

## What is Panchanga?

**Pancha** = Five, **Anga** = Limbs

The five fundamental divisions of time that form the **foundation of any horoscope**. Panchanga must be examined **before** looking at planetary positions, as it shows the **quality of time** for that birth.

### The Five Limbs:

1. **Vara** (Day) - Agni Tatva
2. **Tithi** (Lunar day) - Jala Tatva  
3. **Karana** (Half-tithi) - Prithvi Tatva
4. **Nakshatra** (Lunar mansion) - Vayu Tatva
5. **Yoga** (Sun-Moon combination) - Akasha Tatva

## Benefits of Learning Panchanga

- **Vara** - Prolongs life
- **Tithi** - Brings prosperity
- **Karana** - Gives success in all tasks
- **Nakshatra** - Removes sin
- **Yoga** - Immunity from disease

## 1. VARA (The Day)

**Connection**: Agni (fire), energy levels, physical health

**Calculation**: Sunrise to next sunrise

**Seven Varas**: Ravivara (Sunday), Somavara (Monday), Mangalavara (Tuesday), Buddhavara (Wednesday), Guruvara (Thursday), Shukravara (Friday), Shanivara (Saturday)

**Analysis**: 
- Check **Varesha** (lord of the day) placement from **Lagna**
- Also check Varesha placement from **Surya**

**Example Given**: Franklin D. Roosevelt
- Born Monday (Varesha = Moon)
- Moon in 10th house conjunct Mars
- Moon has **zero Digbala** (directional strength) in 10th
- Moon-Mars = Gandanta (fire-water conflict)
- Moon in 6th from Sun (enemy position)
- Result: Polio (health issues related to movement on 4th-10th axis)

### Vara Chakra Concepts:
- **6th graha** = enemy (Shatru)
- **8th graha** = eclipse (Grahana) - blocks manifestation

## 2. TITHI (Lunar Day)

**Connection**: Jala Tatva, relationships, emotional content, interactions

**Calculation**: Sun-Moon relationship; 1 tithi = 12 degrees of zodiac

**30 Tithis total**: 15 in Krishna Paksha (waning), 15 in Shukla Paksha (waxing)

**Eight Lords**: Surya, Chandra, Mangala, Budha, Guru, Shukra, Shani, Rahu (Ketu has no lordship)

**Analysis**: Check Tithi lord placement from **4th house**

### Problematic Tithis:

**Rikta Tithis** (4th, 9th, 14th) - Most problematic:
- Too much emotion, self-centered
- **Chaturthi** - Ganesha (deity)
- **Navami** - Durga (deity)
- **Chaturdashi** - Shiva (deity)

**Other Doshas**:
- **Ekadashi** (11th) - Addiction/anger issues (Mars/Agni)  fasting advised
- **Trayodashi** (13th) - Wealth problems
- **Krishna Chaturdashi** & **Shukla Pratipada** - Serious relationship issues (related to Soma's sin with Tara)

**Example Given**: Marilyn Monroe
- Tithi lord = Venus
- Venus in 10th house in Aries (enemy sign)
- Zero Digbala
- Had relationships but they were problematic

## 3. DIG CHAKRA (Directional Strength)

Eight directions where grahas have **full potential**:

- **East** (Indra) - **Surya** - Self-realization
- **Southeast** (Agni) - **Shukra** - Details, art
- **South** (Yama) - **Mangala** - Warrior energy
- **Southwest** (Niriti) - **Rahu** - Chaos
- **West** (Varuna) - **Shani** - Detachment
- **Northwest** (Vayu) - **Chandra** - Movement
- **North** (Kubera) - **Budha** - Wealth, profession
- **Northeast** (Ishana) - **Guru** - Knowledge

Plus: **Upward** (Ketu/Ananta - Moksha), **Downward** (Lagna/Brahma - Rebirth)

## 4. KALA CHAKRA (Time/Blockage Wheel)

Shows which grahas **block** each direction:

- Mangala blocks Shukra (Agni corner)
- Guru blocks Mangala (Yama corner)
- Budha blocks Rahu (Niriti corner)
- Shukra blocks Shani (Varuna corner)
- Chandra blocks Budha (Kubera corner)
- Rahu blocks Guru (Ishana corner)
- Ketu blocks Surya (Indra corner)

**Key Point**: These are **pairs** that create either debate (Sattvic) or war (Tamasic)

## Important Reminders:

- Panchanga is the **first layer** - more factors will be added
- Cannot make final judgments from Panchanga alone
- Must learn grahas, rashis, bhavas before complete analysis
- Return to Aramba concepts throughout the year as understanding deepens
